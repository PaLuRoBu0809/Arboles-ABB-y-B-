```python
import pickle  # Librería para guardar/cargar objetos complejos en archivos binarios

class Estudiante:
    def __init__(self, id, nombre, telefono, promedio):
        # Constructor: Define las propiedades básicas de cada alumno
        self.id = id
        self.nombre = nombre
        self.telefono = telefono
        self.promedio = promedio

    def __repr__(self):
        # Formato de impresión: Define cómo se verá el objeto al hacer 'print(estudiante)'
        return f"ID: {self.id} | Nombre: {self.nombre}| Tel: {self.telefono} | Promedio: {self.promedio}"


def guardar_datos(arbol, nombre_archivo):
    # 'wb' abre el archivo para escritura binaria
    with open(nombre_archivo, 'wb') as f:
        pickle.dump(arbol, f)  # Serializa el árbol completo y lo guarda en el disco
    print("Archivo guardado.")

def cargar_datos(nombre_archivo):
    # 'rb' abre el archivo para lectura binaria
    with open(nombre_archivo, 'rb') as f:
        # Recupera el objeto (el árbol con todos sus nodos) tal como estaba
        return pickle.load(f)
```

```python
class NodoABB:
    def __init__(self, clave):
        self.val = clave
        self.izq = None
        self.der = None

class ABB:
    def __init__(self):
        self.raiz = None

    def insertar(self, estudiante): # Cambiamos 'clave' por 'estudiante' para mayor claridad
       if self.raiz is None:
            self.raiz = NodoABB(estudiante)
       else:
         # Aquí solo pasamos dos cosas: el nodo actual y el objeto
        self._insertar_recursivo(self.raiz, estudiante)

    def _insertar_recursivo(self, actual, estudiante):
    # Ya no pedimos 'id' como argumento extra, lo sacamos del objeto
      if estudiante.id < actual.val.id:
          if actual.izq is None:
              actual.izq = NodoABB(estudiante)
          else:
              self._insertar_recursivo(actual.izq, estudiante)
      elif estudiante.id > actual.val.id:
          if actual.der is None:
              actual.der = NodoABB(estudiante)
          else:
              self._insertar_recursivo(actual.der, estudiante)

    def buscar(self, id_buscado, actual="raiz"):
      if actual == "raiz":
        actual = self.raiz

    # 1. Si llegamos al final y no está, o si el ID coincide
      if actual is None:
          return None

      if actual.val.id == id_buscado: # Comparamos el ID del objeto
          return actual.val

      # 2. Decidimos el camino usando el ID del objeto guardado
      if id_buscado < actual.val.id:
          return self.buscar(id_buscado, actual.izq)
      else:
          return self.buscar(id_buscado, actual.der)

    def mostrar_paginado(self):
        if not self.raiz:
            print("Árbol vacío.")
            return

        pila = [self.raiz]
        i = 0

        while pila:
            n = pila.pop()
            i += 1

            print(n.val)

            if n.der: pila.append(n.der)
            if n.izq: pila.append(n.izq)

            # La pausa cada 1000
            if i % 1000 == 0:
                input(f"\n--- Se han mostrado {i} registros. Presiona ENTER para los siguientes 1000 ---")
```

```python
class ArbolBPlus:
    def __init__(self, orden=3):
        # Inicializa la raíz como hoja y el orden del árbol
        self.raiz = NodoBPlus(True)
        self.orden = orden
        self.ultimo_id = 0 # Control para IDs automáticos

    def buscar(self, id_buscado):
        # Navega por los nodos internos hasta llegar a la hoja correcta
        n = self.raiz
        while not n.es_hoja:
            i = 0
            while i < len(n.claves) and id_buscado >= n.claves[i]: i += 1
            n = n.hijos[i]
        # Retorna el estudiante si el ID coincide
        return next((e for e in n.claves if e.id == id_buscado), None)

    def insertar(self, est):
        # Si la raíz se llena, el árbol crece en altura (se divide)
        if len(self.raiz.claves) == self.orden:
            nueva = NodoBPlus()
            nueva.hijos.append(self.raiz)
            self._dividir(nueva, 0)
            self.raiz = nueva
        self._insertar_no_lleno(self.raiz, est)

    def _dividir(self, padre, i):
        # Divide un nodo lleno en dos y sube la clave media al padre
        lleno, mid = padre.hijos[i], self.orden // 2
        nuevo = NodoBPlus(lleno.es_hoja)
        punto = lleno.claves[mid].id if lleno.es_hoja else lleno.claves[mid]

        padre.claves.insert(i, punto)
        padre.hijos.insert(i + 1, nuevo)

        if lleno.es_hoja:
            nuevo.claves, lleno.claves = lleno.claves[mid:], lleno.claves[:mid]
            # Mantiene la conexión entre hojas para el listado ordenado
            nuevo.siguiente, lleno.siguiente = lleno.siguiente, nuevo
        else:
            nuevo.claves, lleno.claves = lleno.claves[mid+1:], lleno.claves[:mid]
            nuevo.hijos, lleno.hijos = lleno.hijos[mid+1:], lleno.hijos[:mid+1]

    def _insertar_no_lleno(self, n, est):
        # Inserta el dato en un nodo que tiene espacio disponible
        i = 0
        while i < len(n.claves) and est.id > (n.claves[i].id if n.es_hoja else n.claves[i]):
            i += 1

        if n.es_hoja:
            n.claves.insert(i, est)
        else:
            if len(n.hijos[i].claves) == self.orden:
                self._dividir(n, i)
                if est.id > n.claves[i]: i += 1
            self._insertar_no_lleno(n.hijos[i], est)

    def imprimir_paginado(self):
        # Recorrido lineal a través de las hojas usando el puntero 'siguiente'
        n = self.raiz
        if not n.claves: return print("Árbol vacío")
        while not n.es_hoja: n = n.hijos[0]

        count = 0
        while n:
            for c in n.claves:
                count += 1
                print(f"[{count}] {c}")
                if count % 1000 == 0:
                    input("\n--- ENTER para los siguientes 1000 ---")
            n = n.siguiente
```

```python
import random

def generar_10k_estudiantes(arbol_abb, arbol_bplus):
    # Lista de nombres base para elegir al azar
    nombres = ["Andres", "Beatriz", "Carlos", "Diana", "Esteban", "Fabiola", "Gustavo"]
    # Generamos 10,000 IDs únicos
    ids = list(range(10000))
    #la lista para que los IDs entren en orden aleatorio.
    random.shuffle(ids)

    for i in ids:
       # Creamos un objeto de la clase Estudiante
        nuevo = Estudiante(
            id=i,
            # Elige un nombre al azar de la lista
            nombre=f"{random.choice(nombres)}",
            # Genera un número de teléfono ficticio de 9 dígitos
            telefono=str(random.randint(300000000, 399999999)),
            # Genera un promedio entre 0.0 y 5.0 con un decimal
            promedio=round(random.uniform(0.0, 5.0), 1)
        )
        # El ABB lo organiza de forma jerárquica
        arbol_abb.insertar(nuevo)
        # El B+ lo organiza manteniendo las hojas balanceadas y enlazadas
        arbol_bplus.insertar(nuevo)
    arbol_bplus.ultimo_id = 10000
    print("10,000 estudiantes cargados con éxito")
```

```python

def menu():
    # Instanciamos ambos árboles para el proyecto
    mi_abb = ABB()
    # El Árbol B+ se crea con un orden de 50 (capacidad de nodos).
    # Es ideal para bases de datos y listados ordenados.
    mi_bplus = ArbolBPlus(orden=50)
    # BUCLE PRINCIPAL
    # Usamos 'while True' para que el programa no se cierre hasta que el usuario elija "Salir".
    while True:
        print("\n--- MOTOR DE BÚSQUEDA (ABB & B+) ---")
        print("0.cargar 10000 estudiantes")
        print("1. Insertar valor")
        print("2. Buscar valor")
        print("3. Salir")
        print("4. listar ordenado")
        print("5. listar desordenado")

        opcion = input("Selecciona una opción: ")
        # Llama a la función externa que genera datos aleatorios y los inserta en ambos árboles.
        if opcion == "0":
            generar_10k_estudiantes(mi_abb, mi_bplus)

        elif opcion == "1":
          # El ID se autoincrementa basándose en el último ID registrado en el Árbol B+.
            mi_bplus.ultimo_id += 1
            nuevo_id = mi_bplus.ultimo_id

            print(f"Asignando ID automático: {nuevo_id}")
            nom_in = input("Nombre: ")
            tel_in = input("Tel: ")
            prom_in = float(input("Promedio: "))

            # Creamos el objeto de la clase Estudiante con los datos capturados.
            nuevo = Estudiante(nuevo_id, nom_in, tel_in, prom_in)


            mi_bplus.insertar(nuevo)
            print(f"Valor {nuevo} insertado en el B+.")

            mi_abb.insertar(nuevo)
            print(f"Valor {nuevo} insertado en el ABB.")

        elif opcion == "2":
            id_buscado = int(input("¿Qué número id buscas?: "))
            #buscamos en el abb
            encontrado_abb = mi_abb.buscar(id_buscado)
            # Buscamos en el B+
            encontrado_bplus = mi_bplus.buscar(id_buscado)

            print(f"[ABB]  ¿Encontrado?: {encontrado_abb}")
            print(f"[B+]  ¿Encontrado?: {encontrado_bplus}")

        elif opcion == "3":
            print("Cerrando")
            break
        elif opcion == "4":
            # El Árbol B+ es superior para listar porque sus hojas están conectadas.
            mi_bplus.imprimir_paginado()

        elif opcion == "5":
          # El ABB muestra los datos según cómo fueron insertados (recorrido de árbol).
            # Como los 10k se insertaron aleatoriamente, el listado saldrá "desordenado".
            mi_abb.mostrar_paginado()
        else:
            print("Opción no válida, intenta de nuevo.")

menu()
```
